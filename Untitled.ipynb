```python
def read_file(filename):
    with open(filename) as f:
        lines = []
        cur_lines = [1]
        for line in f:
            if '-----' in line:
                lines.append(cur_lines)
                cur_lines = []
            else:
                cur_lines.append(line)
    return lines[1:]
```
